```python
from langchain_community.retrievers import WikipediaRetriever
```

```python
# now have to make a object of wikipedia retriever
# Initialization the retriever (optional: set language and top_k)

# 1. top_k_results ; how many documents you want in the return 
# 2. lang ;; in which language you want your result

retriever= WikipediaRetriever(top_k_results=2, lang='en') # retriever obj
```

```python
# Define your query

query= 'the geopolitical history of india and pakistan from the perspective of a chinese'

# Get relevent wikipedia documents

docs= retriever.invoke(query) # runnable
```

```python
# print retrieved content

for i, doc in enumerate(docs):
    print(f"/n--- Result {i+1} --")
    print(f"Content:/n{doc.page_content}..") # truncate for display
```
