```python
import random

class NakliLLM:
    def __init__(self):
        print('LLM created')
    
    def predict(self, prompt):

        response_list= [
            'Delhi is the capital of India',
            'IPL is a cricket league',
            'AI stands for Artificial Intellige'
        ]
    
        return {"response": random.choice(response_list)}
```

```python
llm= NakliLLM()
```

```python
llm.predict('What is the capital of India?')
```

```python
class NakliPromptTemplate:

    def __init__(self, template, input_variable):
        self.template= template
        self.input_variable= input_variable

    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
template= NakliPromptTemplate(
    template='write a {length} length poem about {topic}',
    input_variable=['length', 'topic']
)
```

```python
prompt= template.format({'length':'short', 'topic':'india'})
prompt
```

```python
 llm= NakliLLM()
```

```python
llm.predict(prompt)
```

```python
# now try to make a chain to this common step (prompt --> llm) will become easier and direct to use

class NakliLLMChain:
    def __init__(self, llm, prompt):
        self.llm= llm 
        self.prompt= prompt

    def run(self, input_dict):
        prompt= self.prompt.format(input_dict)
        output= self.llm.predict(prompt)

        return output['response']
```

```python
chain= NakliLLMChain(llm, template)
print(chain.run({'length':'long', 'topic': 'delhi'}))
```
